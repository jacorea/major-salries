### Lowest Risk Majors

A low-risk major is a degree where there is a small difference between the lowest and highest salaries. In other words, if the difference between the 10th percentile and the 90th percentile earnings of your major is small, then you can be more certain about your salary after you graduate.

How would we calculate the difference between the earnings of the 10th and 90th percentile? Well, Pandas allows us to do simple arithmetic with entire columns, so all we need to do is take the difference between the two columns:

```python
# calculate the 90-10 percentile salary difference
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
# insert into clean_df with new column called "Spread"
clean_df.insert(1,'Spread', spread_col)
# check insert
clean_df.head()
```

```python
# find top 5 Low risk majors
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major','Spread']].head()
```

```python
# Majors with the highest potential (top 5 in 90th percentile salary range)
high_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_potential[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

```python
# Majors with the greatest spread in salaries
big_sal_diff = clean_df.sort_values('Spread', ascending=False)
big_sal_diff[['Undergraduate Major', 'Spread','Mid-Career 90th Percentile Salary','Mid-Career 10th Percentile Salary']].head()
```

```python
# Majors with highest median salaries
big_med_salaries = clean_df.sort_values('Mid-Career Median Salary', ascending=False)
big_med_salaries[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```

```python
# What categories of degrees have the highest average salaries
# formatting
pd.options.display.float_format = '{:,.2f}'.format
average_salary = clean_df.groupby('Group').mean()
average_salary
```
